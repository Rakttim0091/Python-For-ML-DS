# Bar plot in seaborn library

```python
 df1 = sns.load_dataset("penguins")
```

```python
df1
```

```python
sns.barplot(x = df1.island,y = df1.bill_length_mm)
plt.show()
```

```python
sns.barplot(x = "island", y = "bill_length_mm",data = df1)

plt.show()
```

```python
sns.barplot(x = "island", y = "bill_length_mm",data = df1, hue = "sex")

plt.show()
```

# orider perameter

```python
order1 = ["Dream","Torgersen","Biscoe"]


sns.barplot(x = "island", y = "bill_length_mm", data = df1,
           hue = "sex", order = order1)
plt.show()
```

# hue order

```python
order1 = ["Dream","Torgersen","Biscoe"]



sns.barplot(x = "island", y = "bill_length_mm", data = df1,
           hue = "sex", order = order1, hue_order = ['Female','Male'])
plt.show()
```

```python
order1 = ["Dream","Torgersen","Biscoe"]



sns.barplot(x = "island", y = "bill_length_mm",data = df1,
           color = "red")

plt.show()
```

```python
sns.barplot(x = "island",y = "bill_length_mm",
            data = df1,hue = "sex",palette = "Accent")

plt.show()
```

# Saturation perameter

```python
sns.barplot(x = "island",y = "bill_length_mm",
            data = df1,hue = "sex",
            palette = "gist_heat_r", saturation = 1)

plt.show()
```

```python
sns.barplot(x = "island",y = "bill_length_mm",
            data = df1,hue = "sex",
            palette = "gist_heat_r", saturation = .5)

plt.show()
```

# bins perameter

```python
sns.displot(df1['flipper_length_mm'], bins = [170,180,190,200,210,220,230,240])

plt.show()
```

```python
sns.displot(df1['flipper_length_mm'], 
            bins = [170,180,190,200,210,
                    220,230,240],color = 'red')

plt.show()
```

# KDE

```python
sns.displot(df1['flipper_length_mm'], 
            bins = [170,180,190,200,210,
                    220,230,240], kde = True , color = 'g')

plt.show()
```

```python
sns.displot(df1['flipper_length_mm'], 
            bins = [170,180,190,200,210,
                    220,230,240], kde = False , color = 'g')

plt.show()
```
