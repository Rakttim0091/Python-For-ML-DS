# Linear Graph

```python
x= [1,2,3,4]
y = [5,6,7,8]

plt.plot(x,y)
plt.show()
```

```python
x= [1,2,3,4]
y = [5,6,7,8]
c = 'r'
plt.plot(x,y,c)
plt.show()
```

```python
x = [5,3,4,6,7,8]
y = [3,5,6,7,8,9]
plt.plot(x,y)
plt.show()
```

```python
x = [5,3,4,6,7,8]
y = [3,5,6,7,8,9]

plt.plot(x,y, marker = 'o')
plt.show()
```

```python
x = [5,3,4,6,7,8]
y = [3,5,6,7,8,9]

plt.plot(x,y)
plt.plot(y, marker = ">")
plt.plot(x,marker = "o")
plt.show()
```

# Bar Graph

```python
x = [2,3,4,6,6,9]
y = [5,6,3,9,7,8]
c = ['r','b','y']
plt.bar(x,y,color = c)
plt.show()
```

# Scatter plot

```python
x = [1,2,3,4,5,6]
y = [4,5,2,4,7,8]

plt.scatter(x,y)
plt.show()
```

```python
x = [1,2,3,4,5,6]
y = [4,5,2,4,7,8]
plt.title("Scatter Plot")
plt.xlabel("month")
plt.ylabel('Number')
plt.scatter(x,y)
plt.show()
```

```python
x = [1,2,3,4,5,6]
y = [4,5,2,4,7,8]
plt.title("Scatter Plot")
plt.xlabel("month")
plt.ylabel('Number')

c = ['r','magenta','green','orange','blue','brown']

plt.scatter(x,y, color = c)
plt.show()
```

```python
x = [1,2,3,4,5,6]
y = [4,5,2,4,7,8]
plt.title("Scatter Plot")
plt.xlabel("month")
plt.ylabel('Number')

c = ['r','magenta','green','orange','blue','brown']

plt.scatter(x,y, color = c, s = 150)
plt.show()
```
